```python
import torch
```

```python
import urllib.request

url = 'https://raw.githubusercontent.com/karpathy/makemore/master/names.txt'
filename = 'names.txt'

# This downloads the file
urllib.request.urlretrieve(url, filename)
print(f"Successfully downloaded {filename}")
```

```python
words = open('names.txt', 'r').read().splitlines()

print(words[:10])
print(len(words))
```

```python
min_len = min(len(w) for w in words)
max_len = max(len(w) for w in words)

print("Shortest:", min_len)
print("Longest:", max_len)
```

```python
chars = sorted(list(set(''.join(words))))
print(chars)
print(len(chars))
```

```python
stoi = {ch:i+1 for i, ch in enumerate(chars)}
stoi['.'] = 0

itos = {i:ch for ch,i in stoi.items()}

print(stoi)
```

```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

```python
for w in words:
    chars = ['.'] + list(w) + ['.']
    
    for ch1, ch2 in zip(chars, chars[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
print(N)
```

```python
P = N.float()
P = P / P.sum(dim=1, keepdim=True)
```

```python
print(P)
```

```python
g = torch.Generator().manual_seed(42)
```

```python
ix = 0
out = []

while True:
    p = P[ix]
    ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    
    if ix == 0:
        break
    
    out.append(itos[ix])

print(''.join(out))
```

```python
for i in range(10):
    ix = 0
    out = []

    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        
        if ix == 0:
            break
        
        out.append(itos[ix])

    print(''.join(out))
```

```python
def generate_from_char(start_char):
    # convert input char → index
    ix = stoi[start_char]
    
    out = [start_char]  # start with given char
    
    while True:
        # get probabilities for next char
        p = P[ix]
        
        # sample next char
        ix = torch.multinomial(p, num_samples=1).item()
        
        # stop if "."
        if ix == 0:
            break
        
        # convert index → char
        next_char = itos[ix]
        out.append(next_char)
    
    return ''.join(out)
```

```python
print(generate_from_char('m'))
print(generate_from_char('a'))
print(generate_from_char('s'))
```
